```python
#Importando o Pandas, Importando o Excel 
#Importando o Matplotlib para criar gráfico
#skiprows -> principais linhas a serem removidas
#sep -> separador (Tem CSV que é ; outros | e etc)
#encoding= ->  é um tipo de codificação binária que aceita acentuação

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

dataset = pd.read_csv('Unidades_da_Rede_Federal_de_EPCT.csv',skiprows=0, sep=',', encoding='utf-8')
display(dataset)



# ==================================
#   SE TIVESSE LINHA EM BRANCO
# ==================================

#dataset = dataset.dropna(how='all')
#display(dataset)

#observação
#se e quisesse remover a linha que tiver qualquer NaN eu colocaria dataset = dataset.dropna(how='any')
```

<h3> Unidade de ensino por estado </h3>

```python
ensino_estado = dataset['sigla_uf_unidade_ensino'].value_counts()
print(ensino_estado)
```

```python
#Tamanho do gráfico
plt.figure(figsize=(15,8))

#Título do Gráfico
plt.title('Unidades de ensino por idade')

#Plotando o Gráfico (com as cores e bordas)
ensino_estado.plot.bar(color='darkslategray', edgecolor='y', linewidth=1)

#Título lareral
plt.ylabel('Qtd de Unidades', fontsize='12')

#salvando gráfico
plt.savefig('Unidade_Estado.png', dpi=600)
```
