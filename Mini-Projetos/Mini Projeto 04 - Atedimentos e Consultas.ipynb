```python
#=======================================================
# Dataset criado com um filtro. Apenas dados do:
# HOSPITAL MATERNO INFANTIL DE BRASILIA - HMIB

dataset_hospital = dataset[dataset['Estabelecimento']=='HOSPITAL MATERNO INFANTIL DE BRASILIA - HMIB'].reset_index()

#EXIBIR AS 3 PRIMEIRAS LINHAS - head(3)
display(dataset_hospital.head(3))

#EXIBIR A ULTIMA LINHA - tail(1)
display(dataset_hospital.tail(1))

```

<h3> Novo dataset, com procedimentos de mamografia no HMIB </h3>

```python
#CRIANDO NOVO DATASET com reset_index
dataset_hmib_mamografia = dataset_hospital[dataset_hospital['Procedimento'].str.contains('Mamografia')].reset_index()

#PRIMEIRAS LINHAS (2)
display(dataset_hmib_mamografia.head(2))

#ÚLTIMA LINHA 
display(dataset_hmib_mamografia.tail(1))
```

```python
#Variável 
grafico_subgrupo =dataset_hmib_mamografia['Procedimento'].value_counts()

#Tamanho do gráfico
plt.figure(figsize=(15,8))

#Título do Gráfico
plt.title('Quantidade por Subgrupo')

#Plotando o Gráfico (com as cores e bordas)
grafico_subgrupo.plot.bar(color=(0.1, 0.1, 0.1, 0.5), edgecolor='darkslategray', linewidth=1,hatch='x',linestyle=':' )

#Título lareral
plt.ylabel('Qtd de Unidades', fontsize='12')



```

<h3> Gráfico de Barra (Horizontal) </h3>

```python
#Variável 
#(ASCENDING= TRUE) -> Ordenanar da maior quantidade para a menor
grafico_subgrupo =dataset_hmib_mamografia['Procedimento'].value_counts(ascending=True)

grafico_subgrupo.plot.barh(color='#819c81', edgecolor=(0,0.3,1,0.6), linewidth=2,hatch='|',linestyle='-' , title=' Procedimentos')
```

```python
#===========================================
procedimento_agrupado = dataset_hmib_mamografia.groupby('Procedimento')
procedimento_agrupado = procedimento_agrupado.count() 

meu_explode = [0, 0.08]
label = dataset_hmib_mamografia['Procedimento']



display(procedimento_agrupado)

fig = plt.figure(figsize =(10, 7)) 
plt.pie(grafico_subgrupo, labels=grafico_subgrupo, startangle = 180, explode=meu_explode,colors = ["#20257c", "#B22222"],  autopct='%1.1f%%',  shadow = True )
 
plt.legend(procedimento_agrupado.index, title = "Procedimentos",title_fontsize = 15)
```
