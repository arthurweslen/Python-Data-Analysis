```python
#Importando o Pandas, Importando o Excel 
#Importando o Matplotlib para criar gráfico
#skiprows -> principais linhas a serem removidas
#sep -> separador (Tem CSV que é ; outros | e etc)
#encoding= ->  é um tipo de codificação binária que aceita acentuação

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


dataset = pd.read_csv('cursos-prouni.csv',skiprows=0, sep=',', encoding='utf-8')

display(dataset.head(1))

```

<h3> Tamanho do dataset, e os tipos das colunas </h3>

```python
print(dataset.dtypes)
print('\n')

#SHAPE (nº de linhas,nº de colunas )
shape = dataset.shape
print(shape)
print('linhas,colunas')
```

<h3> Análise geral do dataset </h3>
<i> Contagem, Média, Desvio Padrão, Percentil, Maior e Menor </i>

```python
round(dataset.describe(),2)
```

<h3> Quantidade de Cursos por Estado </h3>

```python
dataset['uf_busca'].value_counts()
```

<h3> Qual curso com a mensalidade mais alta </h3>

```python
#Condição onde a coluna mensalidade for igual a maior mensalidade, depois eu coloquei algumas colunas

dataset[dataset['mensalidade']==dataset['mensalidade'].max()][['grau','turno','mensalidade','curso_busca','cidade_busca','uf_busca','universidade_nome']]



#=======================================================
#Mensaliade mais barata com todas as colunas do dataset

#dataset[dataset['mensalidade']==dataset['mensalidade'].min()]
```

<h3> Qual curso com a mensalidade mais alta por Estado </h3>

```python
 dataset.groupby('uf_busca')['mensalidade'].max()
```

<h3> Todos os dados das mensalidades mais alta por Estado </h3>

```python
dataset.loc[ dataset.groupby('uf_busca')['mensalidade'].idxmax()]
```

<h3> Cursos com valor entre 500 à 800 reais </h3>

```python

dataset[(dataset['mensalidade']>500) & (dataset['mensalidade']<800)][['grau','turno','mensalidade','curso_busca','cidade_busca','uf_busca','universidade_nome']]
```

<h3> Os 5 cursos mais caros </h3>

```python
dataset.nlargest(5,'mensalidade')
```

<h3> Os 10 cursos mais baratos </h3>

```python
dataset.nsmallest(10,'mensalidade')
```
