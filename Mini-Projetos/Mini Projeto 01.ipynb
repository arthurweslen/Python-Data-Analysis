```python
#Importando o Pandas, Importando o Excel e retirando a primeira linha (era um título)
import pandas as pd
dataset = pd.read_excel('cadastro_pessoas.xlsx',skiprows=1)
display(dataset)
```

```python
# Remove as linnhas que todos os campos estão em branco
dataset = dataset.dropna(how='all')
display(dataset)


#observação
#se e quisesse remover a linha que tiver qualquer NaN eu colocaria dataset = dataset.dropna(how='any')
```

```python
# REMOVER COLUNAS Unnamed: 5 e Unnamed: 6
dataset = dataset.drop(['DFS','ERRO'], axis=1)
display(dataset)
```

```python
# RENOMEAR COLUNAS

dataset = dataset.rename(columns={
    'DF.COD':'COD',
    'DF.Nome':'Nome',
    'DF.Sobrenome':'Sobrenome',
    'DF.Idade':'Idade',
    'DF.Sexo':'Sexo' 
})
display(dataset)
```

```python
dataset = dataset.replace({'Masculino':'M'})
display(dataset)
```
