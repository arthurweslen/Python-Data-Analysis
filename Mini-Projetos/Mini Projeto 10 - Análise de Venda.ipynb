### Importando bibliotecas e carregando o arquivo

```python
# Importando o Pandas e Numpy, Carregando o Excel e retirando a primeira linha (era um título)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

dataset = pd.read_excel('venda_valor.xlsx',skiprows=2)
dataset
```

### Remover Linhas e Colunas em Branco

```python
# Remove as colunas em branco
dataset= dataset.dropna(how='all', axis=1)


# Remove as linhas em branco
dataset= dataset.dropna(how='all', axis=0)
display(dataset)

#observação
#se e quisesse remover a linha que tiver qualquer NaN eu colocaria dataset = dataset.dropna(how='any')
```

### Apagar coluna(s) desnecessárias

```python
dataset = dataset.drop(['ORACLE.NUM'], axis=1)
```

### Resetar Index

```python
dataset = dataset.reset_index()
dataset = dataset.drop(['index'],axis=1)
dataset
```

### Renomear Colunas

```python
# RENOMEAR COLUNAS

dataset = dataset.rename(columns={
    'ORACLE.LOJA':'LOJA',
    'ORACLE.VENDENDOR':'VENDENDOR',
    'ORACLE.SEXO':'SEXO',
    'ORACLE.DATA':'DATA',
    'ORACLE.VENDA':'VENDA',
})
display(dataset)
```

### Substituir Valores

```python
dataset['SEXO'] = dataset['SEXO'].replace({'MASCULINO':'M','FEMININO':'F'})
display(dataset)
```

### Tipo das colunas

```python

dataset.dtypes
#se houvesse necessidade de alterar

# STRING:  dataset['colum'].astype(str)
# INTEIRO: dataset['colum'].astype('int64')
# FLOAT: dataset['colum'].astype('float64')
# DATA: pd.to_datetime(dataset['colum'])
```

```python
dataset
```

### Criar coluna Mês e Dia e concatenar

```python
#CRIAR COLUNAS DIA E MES
dataset['DIA'] = dataset['DATA'].dt.day
dataset['MES'] = dataset['DATA'].dt.month

#CONCATENAR E TRANSFORMANDO EM UMA STRING
dataset['DIA/MES'] = dataset['DIA'].map(str) + dataset['MES'].map(str)

#CONVERTER PARA NÚMERO
dataset['DIA/MES'] = dataset['DIA/MES'].astype('int64')
dataset.dtypes
```

### Coluna Estação
<i><b> contar a freqüência (groupby)  </b></i>

```python
dataset['ESTAÇÃO'] = np.where((dataset['DIA/MES']>=213) & (dataset['DIA/MES']<=216), 'Outono',
                              np.where((dataset['DIA/MES']>=226) & (dataset['DIA/MES']<=239),'Inverno',
                                       np.where((dataset['DIA/MES']>=249) & (dataset['DIA/MES']<=2112),'Primavera','Verão')))


contagem_por_estacao = dataset.groupby(['ESTAÇÃO']).count()
contagem_por_estacao
```

### Mudando o formato da data (dd/mm/yyyy)

```python
dataset['DATA'] = dataset['DATA'].dt.strftime('%m/%d/%Y')
```

```python
dataset
```

### Filtrar lojas 1, 4 e 5
<i><b> contar a freqüência (groupby)  </b></i>

```python
# Filtrar lojas
dataset = dataset.loc[
    (dataset['LOJA']==1.0) |
    (dataset['LOJA']==4.0) |
    (dataset['LOJA']==5.0)]

contagem_por_loja = dataset.groupby(['LOJA']).count()
contagem_por_loja
```

```python
total_vendas_por_vendedor = dataset.groupby(['VENDENDOR']).sum()['VENDA']
maiores_vendas_por_vendedor = dataset.groupby(['VENDENDOR']).max()['VENDA']
menores_vendas_por_vendedor = dataset.groupby(['VENDENDOR']).min()['VENDA']
media_vendas_por_vendedor = dataset.groupby(['VENDENDOR']).mean()['VENDA']
```

### Gráfico de Pizza - Total de Vendas por Sexo

```python
contagem_por_sexo = dataset.groupby(['SEXO']).sum()
contagem_por_sexo
```

```python

#CRIAÇÃO DE VARIÁVEIS
contagem_por_sexo = dataset.groupby(['SEXO']).sum()
explode= (0.04,0)
colors = ["#DB7093","#4169E1"]


plt.figure(figsize=(15,8))
plt.title('Total de Vendas por Sexo ', fontsize='14' )


plt.pie(x=contagem_por_sexo['VENDA'], 
        labels = contagem_por_sexo.index,
        explode = explode,
        colors = colors ,
        autopct='%1.1f%%',
        wedgeprops={'edgecolor': 'silver'},
        startangle=90, 
        shadow=True,
        textprops = {"fontsize": 12,
                     "color":'#333333',
                     "backgroundcolor":'#d3d3d3',
                    "alpha":0.8})  

plt.legend(bbox_to_anchor=(1.17, 0.95),
           ncol=1,
           shadow=True,
           fontsize=12,
           facecolor='#f7f1e3',
           edgecolor= 'silver')
    

          
plt.show()
```

### Gráfico de Barra - Total de Vendas por Loja

```python
soma_por_loja = dataset.groupby(['LOJA']).sum()
soma_por_loja['LOJA'] = soma_por_loja.index
soma_por_loja['VENDA'] = soma_por_loja.round(2)
soma_por_loja
```

```python
#CONFIGURAÇÃO PADRÃO DO GRÁFICO
plt.figure(figsize=(15,8))
plt.title('Total de vendas por loja', fontsize='14', fontweight="bold")
plt.ylabel('Valor de Vendas', fontsize='12', fontweight="bold")
plt.xlabel('Lojas', fontsize='12', fontweight="bold")
sns.set_theme(style="dark")


#GRÁFICO DE BARRA
grafico = sns.barplot(data = soma_por_loja,
            x = 'LOJA',
            y = 'VENDA',
            edgecolor='silver',
            palette="Blues_d",
            linewidth=2,
            hatch='.|',
                     
             order=[4.0, 5.0,1.0],
          )

#RÓTULOS
for i in grafico.patches:
    grafico.annotate(i.get_height(),
                  (i.get_x() + i.get_width() / 2, i.get_height()),
                  ha='center',
                  va='baseline',
                  fontsize=13,
                    fontweight="bold",
                  color='black',
                  xytext=(0, 4),
                  textcoords='offset points')
    

soma_por_loja
```
