<h1> Operações Acumulada </h1>
<br>
<i>
• Importar biblioteca pandas <br>
• Importar um arquivo Excel<br>
• Criar coluna com Soma Acumulada - <b> cumsum()</b> <br>
• Criar coluna com Máximo Acumulado - <b> cummax()</b> <br>   
• Criar coluna com Mínimo Acumulado - <b> cummin()</b> <br>
• Criar coluna multiplicando linhas anteriores  - <b> cumprod()</b> <br>

```python
import pandas as pd
dataset = pd.read_excel('lucro_mensal.xlsx')
display(dataset)
```

<h3> Soma Acumulada - cumsum()</h3>

```python
#==================================================
# SOMA ACULUMADA da coluna 'Investimento'

dataset['Soma Acumulada'] = dataset['Investimento'].cumsum()
display(dataset)
```

<h3> Parar de somar com valores em branco </h3>
<br>
<i> Por padrão, se não coloca nada, o <b> cumsum já entende que o skipna é True</b> </i>

```python
dataset['Soma Acumulada'] = dataset['Investimento'].cumsum(skipna=False)
display(dataset)
```

<h3> Maior Acumulado - cummax()</h3>

```python
#==================================================
# MAIOR ACULUMADO da coluna 'Investimento'


dataset['Maior Acumulado'] = dataset['Investimento'].cummax()
display(dataset)
```

<h3> Menor Acumulado - cummin()</h3>

```python
#==================================================
# MENOR ACULUMADO da coluna 'Investimento'


dataset['Menor Acumulado'] = dataset['Investimento'].cummin()
display(dataset)
```

<h3> Multiplicar linhas anteriores - cumprod()</h3>

```python
#===============================================
# Multilicar linhas anteiores da coluna 'Qtd'

dataset['Multiplicar linhas anteriores'] = dataset['Qtd'].cumprod()
display(dataset)
```
