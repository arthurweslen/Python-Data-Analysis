<h1> <center>Ordenar dados </center> </h1>

<h3> <i>• Importar Documento Excel</i> </h3>

```python
import pandas as pd
dataset = pd.read_excel('ordenar.xlsx')
display(dataset)
```

<h3> <i>• Ordenar por Idade - Ordem Decrescente</i> </h3>

```python
#Ordenando a IDADE por ordem CRESCENTE
dataset.sort_values('Idade', ascending=False)
```

<h3> <i>• Ordenar por Nome - Ordem Crescente</i> </h3>

```python
#Ordenando a coluna NOME por ordem CRESCENTE 
dataset.sort_values('Nome', ascending=True)
```

<h3> <i>• Ordenar por Idade e Nome - Ordem Crescente</i> </h3>

```python
dataset.sort_values(['Idade','Nome'],ascending=True, inplace=True)
display(dataset)
```
