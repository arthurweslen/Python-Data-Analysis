<h1>Como Remover colunas e Linas</h1>
<br>
<i> 
• Importar biblioteca pandas<br>
• Importar arquivo Excel<br>
• Remover uma única coluna <br>
• Remover todas colunas em branco (NaN) <br>
• Remover Múltiplas colunas<br>
• Remover Linhas em branco (how='any') e (how='all') <br>
• Remover Linha </i> <br>




<br><b> axis = 0:</b> <i> LINHA </i> <br>
<b> axis = 1:</b> <i> COLUNA </i>

```python
#Importando o Pandas
#Excel já tem cabeçalho (skiprows=0)

import pandas as pd
dataset = pd.read_excel('coluna_apagar.xlsx',skiprows=0)
display(dataset)
```

<h3> Remover uma coluna  </h3>


```python
dataset = dataset.drop(['Coluna Apagar'], axis=1)
display(dataset)

#axis = 1 é coluna
#axis = 0 é linha
```

<h3> Removendo todas as colunas em branco (NaN) </h3>
<i> Todas as colunas onde Todas as linhas estão em branco </i>

```python
dataset = dataset.dropna(how='all', axis=1)
display(dataset)
```

<h3> Remover multiplas colunas, utilizando o nome da coluna </h3>

```python
dataset = dataset.drop(['Coluna E','Coluna F','Coluna G','Coluna H','Coluna I'], axis=1)
display(dataset)
```

<h3> Remover Linhas em Branco </h3><br>
<i> dataset.dropna(how='all', axis=0)-> apaga se todas as linhas estiverem sem valor<br>
<i> dataset.dropna(how='any', axis=0)-> apaga qualquer linha que tiver sem valor 

```python
#Remove quando TODAS AS LINHAS estiverem em branco 
dataset = dataset.dropna(how='all', axis=0)
display(dataset)


#================================================
#OBS: Veja que REMOVEU apenas a linha do ÍNDICE 1
```

```python
#Remove quando linha quando qualquer coluna estiver em branco 
dataset = dataset.dropna(how='any', axis=0)
display(dataset)

#================================================
#OBS: agora foram REMOVIDAS as linhas dos indices:
# 5, 6, 7, 11, 17 e 18
```

<h3> Remover Linhas por INDEX </h3>

```python
# Perceba que as linhas do index 0 e 2 foram deletadas
dataset = dataset.drop([0,2], axis=0)
display(dataset)

```

<h3> Remover linhas duplicadas </h3> 

```python
# Removendo linhas duplicadas apenas da COLUNA A

dataset = dataset.drop_duplicates(subset='Coluna A')
display(dataset)
```
