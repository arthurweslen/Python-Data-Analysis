<h1> CONCAT- Unir Linhas e Colunas </h1>
<br>
<i>
• Importar biblioteca pandas <br>
• Importar 2 arquivos Excel<br>
• Unir as duas tabelas <br>
• Unir as duas tabela, ignorando o Index<br>
• Unir as duas tabelas com as colunas em comum, ignorando o Index<br>
• Unir as duas tabelas e exibir o nome do Dataset ou Tabela <br>

```python
import pandas as pd
dataset1 = pd.read_excel('equipamentos_1.xlsx')
dataset2 = pd.read_excel('equipamentos_2.xlsx')

display(dataset1)
display(dataset2)
```

<h3> CONCAT </h3>

```python
#=======================
#  UNIR AS DUAS TABELAS

dataset_concat = pd.concat([dataset1,dataset2])
display(dataset_concat.head(10))


# Unir dessa forma, ele repete os indices e traz todas as colunas, independente se a coluna não tiver em umas das tabelas
```

```python
#===========================================
#  UNIR AS DUAS TABELAS, IGNORANDO O INDEX 

dataset_concat = pd.concat([dataset1,dataset2], ignore_index=True)
display(dataset_concat.head(10))

#Dessa forma, o próprio Pandas ignora o index das tabelas e cria um index novo.
```

```python
#==============================================================================
#  UNIR AS DUAS TABELAS, IGNORANDO O INDEX E TRAZENDO APENAS AS COLUNAS IGUAIS

dataset_concat = pd.concat([dataset1,dataset2], ignore_index=True, join='inner')
display(dataset_concat.head(10))

# join='inner' traz as colunas em comum
# Se não colocar nada ou colocar join='outer' ele traz todas as colunas
# dataset_concat = pd.concat([dataset1,dataset2], ignore_index=True, join='outer')
```

```python
# UNINDO TABELAS E EXIBINDO O NOME DO DATASET OU TABELA

dataset_concat = pd.concat([dataset1,dataset2], keys=['dataset1','dataset2'], names=['Tabela','ID'])
display(dataset_concat.head(10))
```

```python
# UNINDO TABELAS E EXIBINDO O NOME DO DATASET OU TABELA -> axix=1

dataset_concat = pd.concat([dataset1,dataset2], axis=1)
display(dataset_concat.head(10))
```
