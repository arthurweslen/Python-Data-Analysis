<h1>Agrupar - groupby</h1>

<i> 
• Importar biblioteca pandas<br>

• Agrupar por Estado (soma, máximo, mínimo e média)<br>
• Arrendondar quando necessário, utilizando round() <br>    
• Agrupar por Estado e Cidade<br>

```python
#Importando o Pandas
#Excel já tem cabeçalho (skiprows=0)

import pandas as pd
dataset = pd.read_excel('agrupar.xlsx',skiprows=0)
display(dataset)
```

<h3> • Agrupando por Estado </h3>
<br>
<i>
Soma - Maior - Menor - Média 

```python
#SOMANDO TODO DATASET, AGRUPADO POR ESTADO
estado_sum = dataset.groupby('Estado').sum()

#MAIORES VALORES DA COLUNA, AGRUPADO POR ESTADO
estado_max = dataset.groupby('Estado').max()

#MENORES VALORES DA COLUNA, AGRUPADO POR ESTADO
estado_min = dataset.groupby('Estado').min()

#MÉDIA DOS VALORES DA COLUNA, AGRUPADO POR ESTADO
estado_mean = round(dataset.groupby('Estado').mean(),2)

#========================================================
display(estado_sum)
#display(estado_max)
#display(estado_min)
#display(estado_mean)
```

<h3> • Agrupando por Estado e Cidade </h3>
<br>
<i> Depois de agrupar mais de uma coluna, o index fica bagunçado, então devemos usar o <b>reset_index()</b></i>

```python
#SOMANDO TODO DATASET, AGRUPADO POR ESTADO E CIDADE
#RESET
estado_cidade_profissao = dataset.groupby(['Estado','Cidade']).sum().reset_index()
display(estado_cidade_profissao)
```

<h3>Agrupando por Cidade e Estado, Somando a quantidade de homes e mulheres e trazendo a média de idade</h3> 
<br>
<i>df = df.groupby(['coluna'])agg.({'coluna','soma'})<i>
    

```python
# No agrupamento anterior, não fez sentido somar as idades, então devemos utilizar:

agrupamento_agg = round(dataset.groupby(['Estado','Cidade']).agg({'Total Homens':'sum','Total Mulheres':'sum','Idade': 'mean'}),1).reset_index()
display(agrupamento_agg)
```
