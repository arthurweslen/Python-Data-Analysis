<h1> MERGE- Unir Linhas </h1>
<br>
<i>
• Importar biblioteca pandas <br>
• Importar 2 arquivos Excel<br>
• Unir as duas tabelas <br>
• Apenas os dados que estão em comum em ambas tabelas  <br> 
• Junção com todos valores que tem em comum com a tabela da <u>ESQUERDA</u> <br>
• Junção com todos valores que tem em comum com a tabela da <u>DIREITA</u> <br>
• Todos os valores unidos <br>     
    

```python
import pandas as pd
dataset_esquerda = pd.read_excel('equipamentos_1.xlsx')
dataset_direita = pd.read_excel('equipamentos_2.xlsx')

display(dataset_esquerda)
display(dataset_direita)
```

<h3> Apenas os dados que estão em comum em ambas tabelas </h3><br>
<i> Com o Merge, ele entedeu que a coluna EQUIPAMENTO é a coluna que tem em comum</i>

```python
#======================================================
#Todos os equipamentos que estão no dataset1 (esquerda)
#que não tinha no dataset2 (direita) não vieram

dataset_merge = dataset_esquerda.merge(dataset_direita)
display(dataset_merge)

#obs: o código acima, por padrão how='inner'
#exemplo: dataset_merge = dataset1.merge(dataset2 ,how='inner')
```

<h3> Junção com todos valores que tem em comum com a tabela da <u>ESQUERDA</u> </h3>

```python
#======================================================
#Todos os equipamentos que estão no dataset_esquerda
#que não tinha no dataset_direita  não vieram

dataset_left = dataset_esquerda.merge(dataset_direita, how='left')
display(dataset_left)
```

<h3> Junção com todos valores que tem em comum com a tabela da <u>DIREITA</u> </h3>

```python
#======================================================
#Todos os equipamentos que estão no dataset_direita
#que não tinha no dataset_esquerda não vieram

dataset_right = dataset_esquerda.merge(dataset_direita, how='right')
display(dataset_right)
```

<h3> Todos os valores unidos </h3> 

```python
#======================================================
#Todos os valores unidos, independente se tem valor em comum ou não

dataset_outer = dataset_esquerda.merge(dataset_direita, how='outer')
display(dataset_outer)
```

```python
dataset_outer = dataset_esquerda.merge(dataset_direita, right_index=True, left_index=True ,how='outer')
display(dataset_outer)
```
