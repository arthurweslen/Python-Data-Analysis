<h1>Pivot (Transformar colunas em Index e valores em colunas)</h1>
<br>
<i> 
• Importar biblioteca pandas<br>
• Importar arquivo Excel<br>
• Transformar: 'COD' em índice,'Função' em Colunes e 'Total de Vendas' em Valores <br>
• Transformar: 'COD' em índice,'Loja' em Colunes e 'Total de Vendas' em Valores <br><br>
    • <b>Transformar:</b><br>
    'COD' e 'ID' em índice.
    <br>'Função'e 'Loja' em Colunas.
    <br>'Total de Vendas' em Valores <br>
    
substituir os valores NaN para '-'</i>

```python
#Importando o Pandas
#Excel já tem cabeçalho (skiprows=0)

import pandas as pd
dataset = pd.read_excel('pivot.xlsx',skiprows=0)
display(dataset)
```

<h3>PIVOT</h3>

<i> <b>Transformar:</b>
 <br> <b>• 'COD'</b> em índice,
<br><b>• 'Função'</b> em Colunas e 
<br><b>• 'Total de Vendas'</b> em Valores </i>

```python
dataset_funcao = dataset.pivot(index='COD',columns='Função', values='Total de Vendas')
dataset_funcao = dataset.fillna('-')
display(dataset_funcao)
```

<i> <b>Transformar:</b>
 <br> <b>• 'COD'</b> em índice,
<br><b>• 'Loja'</b> em Colunas e 
<br><b>• 'Total de Vendas'</b> em Valores </i>

```python
dataset_loja = dataset.pivot(index='COD',columns='Loja',values='Total de Vendas')
dataset_loja = dataset_loja.fillna('-')
display(dataset_loja)
```

<i> <b>Transformar:</b>
 <br> <b>• 'ID' e 'COD'</b> em índice,
<br><b>• 'Loja' e 'Função'</b> em Colunas e 
<br><b>• 'Total de Vendas'</b> em Valores </i>

```python
dataset_geral = dataset.pivot(index=['ID','COD'],columns=['Loja','Função'],values='Total de Vendas')
dataset_geral = dataset_geral.fillna('-')
display(dataset_geral)
```
