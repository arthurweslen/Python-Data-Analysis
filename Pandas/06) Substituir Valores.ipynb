<h1>Substituir Valores</h1>
<br>
<i> 
• Importar biblioteca pandas<br>
• Importar biblioteca numpy<br>
• Importar arquivo Excel<br>
• Substituir uma palavra que está presente em toda tabela<br>
• Substituir uma string que está presente em toda tabela<br>
• Substituir valor de uma coluna usando Dicionário <br>
• Substituir valor da tabela usando Lista <br>
• Substituição com condição <b>( WHERE X MASK </b>) <br>
• Substituição com valor Anterior e Posterior <b> fillna(method='ffill')</b> e <b>fillna(method='bfill')</b></i>
   



```python
#Importando o Pandas
#Excel já tem cabeçalho (skiprows=0)

import pandas as pd
import numpy as np
dataset = pd.read_excel('substituir_valores.xlsx',skiprows=0)
display(dataset)
```

<h3> Substituir uma palavra que está presente em toda tabela </h3>
<br> 
<i>•Se não informarmos qual columa queremos substituir, o Pandas substituirá o dataframe todo <br>
<i>• Para deixar o valor nulo, foi necessário usar um comando do NUMPY - np.nan

```python
#Substituir uma PALAVRA presente em toda tabela
dataset = dataset.replace('INVÁLIDO',np.nan)
display(dataset)
```

<h3> Substituir uma String que está presente em toda tabela </h3>
<br> 
<i> A String pode ser um caracter, uma palavra ou um texto, no começo, no meio ou no fim de cada "célula"

```python
#Substituir uma STRING presente em toda tabela
dataset = dataset.replace('.year:2022','', regex=True)
display(dataset)
```

<h3> Substituir valor de uma coluna usando Dicionário </h3>
<br>
<i> Colocar a Palavra que deseja substituir e coloca 'dois pontos' <b>:  palavra nova</b> </i>

```python
dataset['Sexo'] = dataset.replace({'Masculino':'M', 'Feminino':'F'})
display(dataset)
```

<h3> Substituir valor da tabela usando Lista </h3>
<br>
<i> Primeiro coloca entre [] todos os itens que deseja substituir, depois entre [] o valor que vai ser substituido

```python
dataset= dataset.replace(['#','_'],['',''],regex=True)
display(dataset)
```

<h3> Substituição com condição </h3>
<br>

<b> WHERE X MASK </b>

```python
#MASK substituir valor com condição, quando for VERDADEIRO

dataset['Salário'] = dataset['Salário'].mask(dataset['Salário'] <= 1200.00, 1870.00)
display(dataset)
```

```python
#WHERE substituir valor com condição, quando for FALSO

dataset['Salário'] = dataset['Salário'].where(dataset['Salário'] > 1900.00, 2010.00)
display(dataset)
```

<h3> Preencher campo vazio com o valor Anterior <h3>
<i>• fillna(method='ffill')</i>

```python
dataset['Sexo'] = dataset['Sexo'].fillna(method='ffill')
display(dataset)
```

```python
<h3> Preencher campo vazio com o valor Posterior <h3>
<i>• fillna(method='bfill')</i>
```

```python
dataset = dataset.fillna(method='bfill')

display(dataset)
```
