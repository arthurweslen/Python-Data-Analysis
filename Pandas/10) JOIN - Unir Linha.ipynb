<h1> JOIN- Unir Linhas </h1>
<br>
<i>
• Importar biblioteca pandas <br>
• Importar 2 arquivos Excel<br>
• Unir as duas tabelas <br>
• Junção com todos valores que tem em comum com a tabela da <u>ESQUERDA</u> <br>
• Junção com todos valores que tem em comum com a tabela da <u>DIREITA</u> <br>
• Apenas os dados que estão em comum em ambas tabelas  <br> 
• Todos os valores unidos <br>     
    

```python
import pandas as pd
dataset1 = pd.read_excel('equipamentos_1.xlsx')
dataset2 = pd.read_excel('equipamentos_2.xlsx')

display(dataset1)
display(dataset2)
```

<h3> Junção com todos valores que tem em comum com a tabela da <u>ESQUERDA</u> </h3>

```python
#======================================================
#Todos os equipamentos que estão no dataset2 (direita)
#que não tinha no dataset1 (esquerda) não vieram

dataset_left = dataset1.join(dataset2.set_index('Equipamento'), on='Equipamento', how='left')
display(dataset_left)
```

<h3> Junção com todos valores que tem em comum com a tabela da <u>DIREITA</u> </h3>

```python
#====================================================
#Todos os equipamentos que estão no dataset1 (esquerda)
#que não tinha no dataset2 (direita) não vieram

dataset_right = dataset1.join(dataset2.set_index('Equipamento'), on='Equipamento', how='right')
display(dataset_right)
```

<h3> Apenas os dados que estão em comum em ambas tabelas </h3>

```python
#===========================================
# Apenas os dados que estão em comum em ambas
# tabelas (baseado na coluna equipamento)

dataset_inner = dataset1.join(dataset2.set_index('Equipamento'), on='Equipamento', how='inner')
display(dataset_inner)
```

<h3>Todos os valores unidos </h3>

```python
#===================================
#Todas os valores unidos, indepedente se tem comum ou não

dataset_inner = dataset1.join(dataset2.set_index('Equipamento'), on='Equipamento', how='outer')
display(dataset_inner)
```
