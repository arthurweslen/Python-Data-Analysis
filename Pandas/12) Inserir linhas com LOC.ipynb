<h1> Criar linhas com LOC </h1>
<br>
<i>
• Importar biblioteca pandas <br>
• Importar um arquivo Excel<br>
• Criar linhas, informando o valor do índice <br>
• Criar linhas, continuando a sequência do índice <br>    


```python
import pandas as pd
dataset = pd.read_excel('codigos_sucos.xlsx')
display(dataset)
```

<h3> Criar linhas, continuando a sequência do índice </h3>

```python
dataset.loc[len(dataset)] =[21502,'Uva e Morango']
dataset.loc[len(dataset)] =[21522,'Morango e Manga']
dataset.loc[len(dataset)] =[21572,'Morango Laranja']
dataset.loc[len(dataset)] =[2162,'Limão, Abacaxi e Hortelã']
display(dataset)
```

<h3> Criar linhas, informando o valor do índice </h3>

```python
dataset.loc[20]= [21302,'Laranja com Limão']
dataset.loc[21]= [21402,'Laranja, Morango e Manga']
display(dataset)
```
