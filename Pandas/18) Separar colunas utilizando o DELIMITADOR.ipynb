# Separar colunas com o DELIMITADOR

```python
import pandas as pd
dataset = pd.read_excel('cidades.xlsx')
dataset
```

### Separar local em 3 colunas (País, Munícipio e Estado) - Delimitador "/"

```python
dataset[['País','Munícipio','Estado']] = dataset['LOCAL'].str.split('/', expand=True)
dataset
```

### Separar local em 2 colunas (País, Munícipio / Estado) - Delimitador "/"


```python
dataset2 = pd.read_excel('cidades.xlsx')
dataset2[['País','Munícipio / Estado']] = dataset2['LOCAL'].str.split('/', expand=True, n=1)
dataset2

# n=1 vai verificar apenas a primeira / , as outras ele vai ignorar
```

### Separar local em 2 colunas (País/Munícipio e Estado) - Delimitador "/"  
<i> Da direita para a esquerda </i>

```python
dataset3 = pd.read_excel('cidades.xlsx')
dataset3[['País /Munícipio','Estado']] = dataset3['LOCAL'].str.rsplit('/', expand=True, n=1)

dataset3= dataset3.replace(['##12','###','##021'],['','',''],regex=True)
dataset3

#rsplit é diferente do split, pois ele começa a verificar o delimitador da direta para a esquerda (último delimitador)
```
