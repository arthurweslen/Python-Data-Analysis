<H1> Filtrar dados com LOC e ILOC </H1>

<i><h3>Importar biblioteca Pandas </h3></i>

```python
import pandas as pd
dataset = pd.read_excel('LOC.xlsx')
display(dataset)
```

<i><h3>Filtrar linhas por intervalo (primeiras linhas) </h3></i>

```python
# Filtrar as primeiras linhas, no caso, as 7 primeiras linhas
dataset = dataset.iloc[0:7]
display(dataset)
```

<i><h3>Filtro, ignorando a primeira linha </h3></i>

```python
#Remover apenas a primeira linha
dataset = dataset.iloc[1:]
display(dataset)
```

<i><h3>Filtro com condição </h3></i>

```python
# Filtrar linhas que são maiores ou igual a 9 
dataset = dataset.loc[
        dataset['Coluna B']>=9]


display(dataset)
```

```python
# Filtrar linhas que são maiores ou igual a 9 E for positivo (+)
dataset = dataset.loc[
        (dataset['Coluna B']>=9) 
        & 
        (dataset['Coluna C']=='+')]

display(dataset)
```
