<h1> Operçõpes Móveis (Média, Soma, Máx e Min) </h1>
<br>
<i>
• Importar biblioteca pandas <br>
• Importar um arquivo Excel<br>
• Média Móvel de 5 em 5 linhas <br>
• Média Móvel de 5 em 5 linhas, com período mínimo = 1 <br>    
• Soma móvel de 3 em 3<br>
• Maior móvel de 3 em 3<br>
• Menor móvel de 3 em 3<br>


```python
import pandas as pd
dataset1 = pd.read_excel('covid.xlsx', sheet_name='Baixada_Santista')
dataset2 = pd.read_excel('covid.xlsx', sheet_name='Mensal_baixada_santista')

display(dataset1)
display(dataset2.head(2))
```

<h3> Média Móvel de 5 em 5 linhas </h3>

```python
dataset1['Média Movel 5'] = dataset1.rolling(window=5).mean()
#dataset1['teste'] = dataset1['Mortes Covid'].rolling(window=5).mean
display(dataset1)
```

<h3> Média Móvel de 5 em 5 linhas, com período mínimo = 1 </h3>
<br>
As primeiras linhas não teria média, mas com período mínimo 1, ele faz com a quantidade de valores que tem.

```python
dataset1['Média Móvel5 * '] = dataset1['Mortes Covid'].rolling(window=5, min_periods=1).mean()
display(round(dataset1,1))
```

```python
dataset1['Média Móvel5 - center '] = dataset1['Mortes Covid'].rolling(window=5,  center=True).mean()
display(round(dataset1,1))
```

<h3>Soma móvel de 3 em 3 </h3>

```python
dataset1['Soma Móvel  3x3'] = dataset1['Mortes Covid'].rolling(window=3, min_periods=1).sum()
display(dataset1)
```

<h3> Maior móvel de 3 em 3 </h3>

```python
dataset1['Maior Móvel  3x3'] = dataset1['Mortes Covid'].rolling(window=3, min_periods=1).max()
display(dataset1)
```

<h3>Menor móvel de 3 em 3 </h3>

```python
dataset1['Menor Móvel  3x3'] = dataset1['Mortes Covid'].rolling(window=3, min_periods=1).min()
display(dataset1)
```
