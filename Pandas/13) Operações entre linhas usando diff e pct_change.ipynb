<h1> Operações entre linhas usando diff e pct_change </h1>
<br>
<i>
• Importar biblioteca pandas <br>
• Importar um arquivo Excel<br>
• Criar coluna com diferença Bimestral, Trimestral e Semestral <br>
• Criar coluna com diferença em Porcentagem (%)  <br>    

```python
import pandas as pd
dataset = pd.read_excel('lucro_mensal.xlsx')
display(dataset)
```

```python
dataset['Diferença mês seguinte'] = dataset['Investimento'].diff()
display(dataset)
```

```python
dataset['Diferença Bimestral'] = dataset['Investimento'].diff(periods=2)
dataset['Diferença Trimestral'] = dataset['Investimento'].diff(periods=3)
dataset['Diferença Semestral'] = dataset['Investimento'].diff(periods=6)
display(dataset)
```

```python
dataset['Diferença Percental'] = dataset['Investimento'].pct_change()
display(dataset)
```
