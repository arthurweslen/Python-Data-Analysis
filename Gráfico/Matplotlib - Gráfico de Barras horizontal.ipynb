```python
# Importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

#Carregando o exel
dataset = pd.read_excel('Rede_Social.xlsx', sheet_name='Rede_Social',skiprows=0)
display(dataset)
```

```python
dataset.dtypes
```

<h3> Criando o X e o Y para o gráfico </h3>



```python
x = dataset['Rede Social']
y = dataset['Nº Usuários']
```

<h3> Gráfico de Barras Horizontal </h3>

```python
#plt.barh(x,y)

plt.barh(x, y)
```

<h3> Gráfico de Linhas - Personalizado </h3>

```python
#TAMANHO GRAFICO       ====> plt.figure(figsize=(x,y)
#TITULO GRÁFICO        ====> plt.title
#TITULO LATERAL        ====> plt.ylabel
#TITULO RODAPÉ         ====> plt.Xlabel
#GRADE                 ====> plt.grid

#COR DA BARRA          ====> color
#TRANSPARÊNCIA         ====> alpha
#COR DA LINHA          ====> edgecolor
#DESENHO               ====> hatch
#ESPESSURA LINHA       ====> linewidth
#ESTILO LINHA          ====> linestyle


#=== Configuração Geral === 
plt.figure(figsize=(15,8))
plt.title('Total de Usuários por Rede Social')
plt.ylabel('Rede Social', fontsize='14')
plt.xlabel('Qtd de Usuários', fontsize='14')
plt.grid(True, alpha=0.2)

grafico = plt.barh(x, y, 
         alpha=0.8, color=['#008080','#008B8B','#20B2AA','#48D1CC','#40E0D0','#00FA9A','#90EE90','#3CB371','#2E8B57','#006400'],
        edgecolor='silver',
         hatch=' .',linestyle =':',linewidth=2)
```
