<h1>Gráfico de Linhas</h1>

<i> 
• Importe o Pandas<br>
• Importe o Matplotlib<br>
• Importe o Seaborn<br>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

dataset = pd.read_csv('ipeadata_BPM6.csv',skiprows=0, sep=',', encoding='utf-8')
display(dataset.head(4))
```

<h3> Criando variáveis X e Y </h3>

```python
x = dataset['Data']
y = dataset['Balanço de pagamentos - ativos de reserva - US$ - Banco Central do Brasil- Balanço de Pagamentos (BPM6) (BCB / BP (BPM6)) - BPAG_AR']
```

<h3> Criando uma função para configuração básica do gráfico <h3>

```python
 def configuracao(grafico1):
    plt.figure(figsize=(15,8))
    plt.title('Balanço de pagamentos - ativos de reserva por ano', fontsize='14')
    plt.ylabel('Balanço de pagamentos - ativos de reserva ', fontsize='14')
    plt.xlabel('Ano', fontsize='14')
    sns.set_theme(style="whitegrid")
    
def configuracao(grafico2):
     
    plt.figure(figsize=(15,8))
    plt.title('Balanço de pagamentos - ativos de reserva por ano', fontsize='14')
    plt.ylabel('Balanço de pagamentos - ativos de reserva ', fontsize='14')
    plt.xlabel('Ano', fontsize='14')
    sns.set_theme(style="darkgrid")  
```

<h3> Modelo 1 </h3>

```python
configuracao('grafico1')
sns.lineplot(data=dataset,
             x=x,
             y=y,
             color='#1E90FF',
             drawstyle='default',
             marker='o', 
             markerfacecolor='#4169E1',
             markersize=8,
             fillstyle='full'
)
```

<h3> Modelo 2 </h3>

```python
configuracao('grafico2')
sns.lineplot(data=dataset,
             x=x,
             y=y,
             color='#4F4F4F',
             
             linestyle='dashdot',
             linewidth=2,
             drawstyle='steps',
             marker='s', 
             markerfacecolor='#363636',
             markersize=13,
             fillstyle='bottom')
```
