Importar a biblioteca Matplotlib <br>
Importar a biblioteca Numpy <br>

```python
#importando bibliotecas

import matplotlib.pyplot as plt
import numpy as np
```

```python
#Determinando o tamanho da fonte do Gráfico
plt.rcParams.update({'font.size':13})
```

<h3> Criando uma serie de dados </h3><br>
<i>
• 4 Profissões, cada um com sálario de junior, pleno e senior

```python
profissao = ['Gerente','Coordenador','Desenvolvedor','Analista']
salario_junior = [7500,5000,3500,1700]
salario_pleno = [9200,8000,7000,2500]
salario_senior = [13500,9500,15000,3200]
```

```python
x = np.arange(len(profissao))
x
```

<h3> Criando um gráfico de barras simples </h3>

```python
fig,ax = plt.subplots()

jr = ax.bar(x,salario_junior)

```

<h3> Criando gráfico de barras agrupado - personalizado </h3>

```python
#Criando uma variável largura, para as barras
largura = 0.23

#Função para colocar OS ROTUOS NAS BARRAS
def autolabel(grupos):
    for i in grupos:
        h = i.get_height()
        ax.annotate('{}'.format(h),
                    xy = (i.get_x()+i.get_width()/2,h),
                    xytext = (0,4),
                    textcoords = 'offset points',
                    ha = 'center')
               
```

```python
#CONFIGURAÇÃO GERAL DO GRÁFICO
fig,ax = plt.subplots(figsize=(20,9))
#plt.grid(True)


#PLOTANDO CADA BARRA DO GRÁFICO
jr = ax.bar(x - 0.2 ,salario_junior, largura, label = 'Junior', color='#34495e',edgecolor='#2c3e50', linewidth=2,hatch='///')
pl = ax.bar(x + 0 ,salario_pleno,largura, label = 'Pleno', color='#3498db',edgecolor='#2980b9', linewidth=2,hatch='|')
sr = ax.bar(x + 0.2 ,salario_senior,largura, label = 'Senior', color='#1abc9c',edgecolor='#16a085', linewidth=2, hatch='x/O.')

#TÍTULO
ax.set_title('Salário por Profissão e por Nível')
ax.set_ylabel('Salário')
ax.set_ylim([0,16000])

#LEGENDAS NO EIXO X
ax.set_xticks(x)
ax.set_xticklabels(profissao)

#RÓTULOS
autolabel(jr)
autolabel(pl)
autolabel(sr)

ax.legend()
plt.show
```
