```python
#Importando o Pandas, Importando o Excel 
#Importando o Matplotlib para criar gráfico
#skiprows - principais linhas a serem removidas

import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_excel('Dollar.xlsx',sheet_name='Dollar', skiprows=0)
dataset
```

```python
ano = dataset['Ano']
americano = dataset['Dolar Americano']
canadense = dataset['Dolar Canadense']

plt.fill_between(ano, americano, label='Dólar Americano')
plt.fill_between(ano, canadense, label='Dólar Canadense')

plt.legend()
```

```python
plt.figure(figsize=(15,8))
plt.grid(True, alpha=0.2)
plt.title('Valor do dólar a cada ano', fontsize='14' )


plt.fill_between(ano, americano,
                 alpha=0.8, 
                 hatch='./.',
                 edgecolor='#27ae60',##3c40c6
                 facecolor='#16a085',
                 linewidth=4,
                 linestyle = '-',
                 label='Dolar Americano')

plt.fill_between(ano, canadense,
                 alpha=0.8,
                 hatch='..',
                 edgecolor='#34495e',##
                 facecolor='#3498db',
                 linewidth=4,
                 linestyle = '-',
                 label='Dolar Canadense'
                )


plt.fill_between(ano, canadense,americano,
                where = (canadense >= americano),
                 alpha=0.8,
                 hatch='..',
                 edgecolor='red',##
                 facecolor='red',
                 linewidth=4,
                 linestyle = '-',
                 label='Dolar canadense mais alto que o americano')


plt.legend(loc="upper left",
           #bbox_to_anchor=(1.17, 0.95),
           ncol=1,
           shadow=True,
           fontsize=12,
           facecolor='#ecf0f1',
           edgecolor= 'yellow'
          )

           
```
