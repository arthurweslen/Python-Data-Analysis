```python
#Importando o Pandas, Importando o Excel 
#Importando o Matplotlib para criar gr치fico
#skiprows - principais linhas a serem removidas

import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_excel('setor.xlsx',sheet_name='Setor', skiprows=0)
dataset
```

```python
#CRIA칂츾O DE VARI츼VEIS
explode= (0.04,0,0,0)
colors = ["#7d5fff", "#18dcff", "#32ff7e","#cd84f1"]


plt.figure(figsize=(15,8))
plt.title('Total de funcion치rios por setor', fontsize='14' )


plt.pie(x=dataset['Funcion치rios'], labels = dataset['Setor'],
        explode = explode,
        colors = colors ,
        autopct='%1.1f%%',
        wedgeprops={'edgecolor': 'yellow'},
        startangle=90, 
        shadow=True,
        textprops = {"fontsize": 12,
                     "color":'#333333',
                     "backgroundcolor":'#d3d3d3',
                    "alpha":0.8}) 

plt.legend(bbox_to_anchor=(1.17, 0.95),
           ncol=1,
           shadow=True,
           fontsize=12,
           facecolor='#f7f1e3',
           edgecolor= 'yellow')
           
          
plt.show() 
```
