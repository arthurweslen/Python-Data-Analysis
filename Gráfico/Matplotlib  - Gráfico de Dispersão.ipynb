```python
#Importando o Pandas, Importando o Excel 
#Importando o Matplotlib para criar gráfico
#skiprows - principais linhas a serem removidas

import matplotlib.pyplot as plt
import numpy as np
import random
```

### Criando variávies para plotar no gráfico

```python
#Números de 0 a 150
x = np.arange(0,151,5)
z = np.arange(0,151,10)


#números aleatórios
y = np.random.random(len(x))
k = np.random.random(len(z))
```

### Gráfico de dispersão 

```python
plt.scatter(x,y)
plt.show()
```

### Gráfico de dispersão personalizado

```python
#CONFIGURAÇÃO GERAL DO GRÁFICO
plt.figure(figsize=(15,8))
plt.title('Gráfico de Dispersão', fontsize='16',fontweight='bold',pad=20 )

plt.xlabel('Eixo X', fontsize='12',fontweight='bold')
plt.ylabel('Eixo Y', fontsize='12',fontweight='bold')
plt.xlim([0,160])
plt.ylim([0,1.2])

plt.grid(True, alpha=0.2)



#====================
plt.scatter(x,y,
            color = '#e056fd',
            marker = "*",
            s = 110 ,#tamano
            edgecolor = '#ffbe76')

plt.scatter(z,k,
            color = '#686de0',
            marker = "P",
            s = 60, #tamano
            edgecolor = '#130f40' 
           )

plt.show()
```
